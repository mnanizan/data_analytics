# 4.1 Object Oriented Interface (OOI) vs Pyplot

```python
%matplotlib inline
```

matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g. creates a figure, creates a plotting area ina figure, plots some lines in a plotting area, decorates the plot with label, etc. 
In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure & plotting area, & the plotting functions are directed to the current subplot.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,2.5])
plt.ylabel('some numbers')
```

plot() is a versatile command, and will take an arbitary number of arguments. For example, to plot x versus y, you can issue the command:

```python
plt.plot([1,2,3,4,],[1,4,9,16]);
```

using the pyplot interphase, you build a graph by calling a sequence of cuntions and all of them are applied to the current subplot, like so:

```python
plt.plot([1,2,3,4],[10,20,25,30], color='lightblue', linewidth=3)
plt.scatter([0.3,3.8,1.2,2.5],[11,25,9,26], color='darkgreen', marker='^')
plt.xlim(0.5, 4,5) # to set the min & max value at x-axis
plt.title('Title of this plot')
plt.ylabel('This is the x-label')
plt.xlabel('This is the y-label')
```

when working with just one subplot in the figure, generally is OK to work with the pyplot interphase, however when doing more complicated plots, or working within larger scripts, you will want to explicitly pass around the the Subplot(Axes) and/or Figure to operate on

```python
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0,0.02)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(t1, f(t1), 'bo') # b.o is for a doted line

plt.ylabel('Y label')

plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.ylabel('Y label')
```

# 4.2 Object Oriented Interface 

What you do when plotting with the object oriented interphase, is to create the objects & then call the methods for every objects to make changes specifically to that objec. The plt.subplots function creates a figure object & an Subplot(Axes) object at the same time, then we can create a plot by calling methods on those objects.

```python
x = np.linspace(-4,4)
y1, y2 = x**2, x**3
```

```python
# Object oriented interface
fig, ax = plt.subplots()
ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
ax.set_title('Quadratic and Cubis Powers')
ax.set_xlabel('x values')
ax.set_ylabel('y values')
```

Interestingly, just about all methods of an Axes object exist as a function in the pyplot module (and vice-versa). For example, when calling plt.xlabel('x label'), pyplot calls ax.set_xlabel('x label')on whichever Axes is 'current'. Here is an equivalent version of the above example using ust pyplot.

```python
# pyploy interface
plt.plot(x, y1, 'red', x, y2, 'blue')
plt.title('Qudaratic & Cubic Powers')
plt.xlabel('x values')
plt.ylabel('y values')
```

The usefulness of the OOI is clearer when working with mutilple subplots in one figure. We've mentioned before that a figure can have more thn one Axes on it.If you want your axes to be on regular grid system, then it's easiest to use plt.subplot(...) to create a figure and add the axes to it automatically.

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
```

plt.subplots(...)created a new figure and added 4 subplots to it. The axes object that was returned is a 2D numpy object array. Each item in the array is one of the subplots. They're laid out as you see them on the figure. Therefore, when we want to work with one of these axes, we can index the axes array and use that items's methods. For example:

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].set_title('Upper Left')
axes[0,1].set_title('Upper Right')
axes[1,0].set_title('Lower Left')
axes[1,1].set_title('Upper Right')

# To iterate over all items in a multidimensional numpy array, use the 'flat' attribute
for ax in axes.flat:
    # Remove tall xticks & yticks
    ax.set(xticks=[], yticks=[])
    
fig.tight_layout();
```

## 4.2.1 Example: Creating a figure showing the shape of the power functions. 

```python
x = np.linspace(start=-5, stop=5, num=150)
```

```python
# All functions in one axes
fig, ax=plt.subplots(figsize=(7,4))
ax.plot(x, x, label='Linear')
ax.plot(x, x**2, label='Quadratic')
ax.plot(x, x**3, label='Cubic')
ax.plot(x, x**4, label='4th Power')
ax.legend();
```

```python
# 4 Axes/Subplots: One fucntion in one axes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7,4.5))
axes[0,0].set_title('Linear')
axes[0,0].plot(x,x)
axes[0,1].set_title('Quadratic')
axes[0,1].plot(x,x**2)
axes[1,0].set_title('Cubic')
axes[1,0].plot(x, x**3)
axes[1,1].set_title('4th Power')
axes[1,1].plot(x, x**4)
fig.tight_layout();
```

What if we want to take a look at teh first 10 powers of x

```python
# A more elegant approach
fig, axes= plt.subplots(nrows=2, ncols=5, figsize=(14,4.5))
for i, ax in enumerate(axes.flatten()):
    ax.set_title('x to the {:d}'.format(i+1))
    ax.plot(x, x**(i+1))
    
fig.tight_layout()
```

# 4.3 Common Customization: colors, labels, ticks, tickmarks, limits & annotations

```python
# Generate data
x = np.linspace(-np.pi, np.pi, 200)
sine, cosine = np.sin(x),np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, sine, x, cosine);
```

to this plot:

```python
fig, ax = plt.subplots()
ax.plot(x, sine, color='red', label='Sine')
ax.plot(x, cosine, color='#165181', label='Cosine')

ax.set_xlim(-3.5,3.5)
ax.set_ylim(-1.2,1.2)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1,0,1])

ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'], size=17)
ax.set_yticklabels(['-1', '0','+1'], size=17)

ax.legend(loc='upper left')

ax.text(-0.25, 0, '(0,0)') #x-coor,y-coor
ax.text(np.pi-0.2, 0.05, r'$(\pi, 0)$', size=15)

ax.annotate('Origin',
           xy=(0,0), #where the arrow points at
           xytext=(1, -0.7),#location of text
           arrowprops=dict(facecolor='blue'));

ax.axhline(0, color='black', alpha=0.9) #horizontal line
ax.axvline(0, color='black', alpha=0.9) # vertical
ax.grid();
```
